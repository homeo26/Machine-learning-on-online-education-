```python
# df = OES.drop(columns=['Device type used to attend classes', 'Your interaction in online mode','Your level of satisfaction in Online Education'])

# Also we can drop columns using the index of each column
df = OES.drop(OES.columns[[4, 5, 18, 19, 21]], axis=1)
df
```

```python
# As a part of exploration of the Data
# We must check for NaN values  

df.isna().sum() 

# df = df.dropna()

# here as we can see there are no NaN values
```

```python
score = 'Average marks scored before pandemic in traditional classroom'
df[score] = df[score].replace({'20-Nov': '11-20'})
df[score] = df[score].replace({'0-10': 5})

for i in range(10,100,10):
  df[score] = df[score].replace({str(i+1)+'-'+str(i+10): int(i+5)})
  
# df.to_csv("OES.csv", index=False)
df
```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object gives uniqe values to the all items.
label_encoder = preprocessing.LabelEncoder() 


li = [0, 9, 10, 14, 15, 16, 22]
for i in li:
  df[OES.columns[i]]= label_encoder.fit_transform(df[OES.columns[i]])
  df[OES.columns[i]].unique()

df['Economic status'] = df['Economic status'].replace('Poor',0)
df['Economic status'] = df['Economic status'].replace('Middle Class',1)
df['Economic status'] = df['Economic status'].replace('Rich',2)

df
```

```python
# Also we can use OneHotEncoder as well but we wanted to specify columns more specificly 

df["Interested in?"] = df["Interested in?"].replace('Theory','1,0')
df["Interested in?"] = df["Interested in?"].replace('Practical','0,1')
df["Interested in?"] = df["Interested in?"].replace('Both','1,1')

df['Level of Education']=df['Level of Education'].replace('School',0)
df['Level of Education']=df['Level of Education'].replace('Under Graduate',1)
df['Level of Education']=df['Level of Education'].replace('Post Graduate',2)

df['Home Location']=df['Home Location'].replace('Rural',0)
df['Home Location']=df['Home Location'].replace('Urban',1)

df
```

```python
# df["Interested in?"].str.split(',', expand=True)

df[['Theory', 'Practical']] = df["Interested in?"].str.split(',', expand=True)
df = df.drop("Interested in?",axis=1)

df['Theory'] = df['Theory'].map(lambda ele: int(ele) )
df['Practical'] = df['Practical'].map(lambda ele: int(ele) )

df
```

```python
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = df.drop(columns = ['avg score'])
Y = df['avg score']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)

model = dtc()
model.fit(X_train,Y_train)
perd = model.predict(X_test)
score = accuracy_score(Y_test, perd)
```
